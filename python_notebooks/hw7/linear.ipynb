```python
import statsmodels.api as sm
import pandas as pd
```

```python
# load data
moral_2 = pd.read_csv('../clean_data/expanded_moral_temp0.2.csv')
moral_7 = pd.read_csv('../clean_data/expanded_moral_temp0.7.csv')
risk_2 = pd.read_csv('../clean_data/expanded_risk_temp0.2.csv')
risk_7 = pd.read_csv('../clean_data/expanded_risk_temp0.7.csv')
```

# Examine Moral Stories

## Moral Stories 0.2

```python
def simple_ols(df, y, x):
    X = df[x]
    y = df[y]
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    return model.summary()
```

```python
simple_ols(moral_2, 'moral_score', "self_ext_total")
```

```python
simple_ols(moral_2, 'moral_score', "self_agr_total")
```

```python
simple_ols(moral_2, 'moral_score', "self_con_total")
```

```python
simple_ols(moral_2, 'moral_score', "self_neu_total")
```

```python
simple_ols(moral_2, 'moral_score', "self_ope_total")
```
